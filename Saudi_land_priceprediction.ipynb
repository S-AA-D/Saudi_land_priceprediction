## Start building the model

### From here you can build a model that predict all lands in the cities of Saudia Arabia but to improve the model accuracy in predicting I will specify only the Riyadh and include them in the training and evaluating

```python
# Skip the data-preprocessing steps below and use final_df if you want to predict base on all cities and nerighbors in KSA

top_cities_df = df[df['المدينة'].isin(['الرياض'])]
top_cities_df.drop(columns = ['السنة' , 'المنطقة'] , inplace = True)
top_cities_df.reset_index(drop = True , inplace = True)
top_cities_df.drop(columns = 'المدينة' , inplace = True)
top_cities_df
```

```python
top_cities_df['الحي'].value_counts()
```

```python
# For الحي feature apply the same logic as city but in this case we will take top 100
top_50_cities = top_cities_df['الحي'].value_counts().nlargest(150)
top_50_cities

top_50_cities_indicis = top_50_cities.index
top_cities_df.loc[~top_cities_df['الحي'].isin(top_50_cities_indicis), 'الحي'] = 'other_nerighbors'
top_cities_df['الحي'].value_counts()
```

```python
categorical_features = [top_cities_df.columns[i] for i in range(len(top_cities_df.columns)) if top_cities_df.dtypes[i] == 'object']
categorical_features
```

```python
encoder = OneHotEncoder(sparse=False)
encoded_features = encoder.fit_transform(top_cities_df[categorical_features])
encoded_top_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))
```

```python
encoded_top_df
```

```python
top_final_df = pd.concat([top_cities_df.drop(columns = categorical_features) , encoded_top_df] , axis =1)
```

```python
top_final_df.describe()
```

```python
#From the earlier data.describe() output, I observed that the values of 2 variables - السعر and المساحة - are large and will affect the absolute numbers of the regression model.
#To manage this, I will normalize the data using log.
top_final_df['السعر'] = np.log(top_final_df['السعر'])
top_final_df['المساحة'] = np.log(top_final_df['المساحة'])
```

```python
top_final_df['المساحة'].sort_values()
```

```python
top_final_df.describe()
```

```python
# Drop infinity values (it's only in المساحة and on ly one feature)
top_final_df.drop(top_final_df[top_final_df['المساحة'] == -np.inf].index , inplace = True)
top_final_df
```

```python
x = top_final_df.drop(columns = 'السعر')
y = top_final_df['السعر']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(x , y , test_size = 0.4 , random_state = 33 ,shuffle=True)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential([
    Dense(256 , activation = 'relu' , input_shape = [170]),# If u used final_df change the input shape to 288
    Dense(128 , activation = 'relu'),
    Dense(64 , activation = 'relu'),
    Dense(1 , activation = 'linear')
])
model.compile(optimizer = 'adam' , loss = 'mse' , metrics = ['mean_absolute_error' , 'mean_squared_error' ])
model.summary()
```

```python
model.fit(X_train , y_train , epochs = 50 , batch_size = 512)
```

```python
model.evaluate(X_test , y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = y_pred.flatten()
y_pred
```

```python
actual_y_test = np.exp(y_test)
actual_predicted = np.exp(y_pred)
compare_actual = pd.DataFrame({'Test Data': actual_y_test, 'Predicted Price' : actual_predicted})
compare_actual['Dfference'] = compare_actual['Test Data'] - compare_actual['Predicted Price']
compare_actual = compare_actual.astype(int)
compare_actual.head(10)
```
